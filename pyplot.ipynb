## <center>Pyplot</center>

## Définition

<strong>Pyplot</strong> est un sous module de matplotlib contenant la plupart des utilitaires (fonctions) dont la librairie matplotlib a besoin pour :
<ul>
    <li><strong>la création de figures </strong></li>
    <li><strong>la création de zones de traçage dans une figure </strong></li>
    <li><strong>tracer certaines lignes dans une zone de traçage </strong></li>
    <li><strong>décorer des tracés avec des étiquettes, etc.</strong></li>
</ul>    
    Ces utilitaires sont généralement importés sous l'aléas plt. 

Avec pyplot, nous pouvons effectuer des tracés de ligne, histogramme, Scatter, tracé 3D, image, contour, etc.

## Installation

### Quelques fonctions de pyplot

```python
# Programme python pour montrer les fonctions de pyplot
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.axis([0, 6, 0, 20])
plt.show()
```

<ul>
   <li><strong>plot(x, y) :</strong> tracez x et y en utilisant le style de ligne et la couleur par défaut.</li>
   <li><strong>plot.axis([xmin, xmax, ymin, ymax])</strong></li>
   <li><strong>plot(x, y, color='green', marker='o', linestyle='dashed', linewidth=2, markersize=12, label='La légende de la courbe')</strong></li>
   <li><strong>plot.xlabel()</strong></li>
   <li><strong>plot.ylabel()</strong></li>
</ul>   

### Exemple1 : Tracé linéaire 

### Ecrivons un programme qui nous permet de visualiser l'évolution de la consommation d'électricité : Cas de la Cote d'Ivoire et du Burkina.

```python
annee = [1972, 1982, 1992, 2002, 2012]
e_cote_ivoire = [100.6, 158.61, 305.54, 394.96, 724.79]
e_burkina = [10.5, 25.21, 58.65, 119.27, 274.87]
  

plt.plot(annee, e_cote_ivoire, color ='orange', 
         label ="Cote d'Ivoire")
  
plt.plot(annee, e_burkina, color ='g', 
         label ='Burkina')
  
# Noms des axes
plt.xlabel('Années')
plt.ylabel("Consommation d'énergie en kwh")
  
# Titre
plt.title("Consommation d'électricité par habitant")
  
plt.legend()
plt.show()
```

### Exemple 2 : Tracé linéaire avec mise en forme des lignes.

```python
annee = [1972, 1982, 1992, 2002, 2012]
e_cote_ivoire = [100.6, 158.61, 305.54, 394.96, 724.79]
e_burkina = [10.5, 25.21, 58.65, 119.27, 274.87]
  

plt.plot(annee, e_cote_ivoire, color ='orange', 
         label ="Cote d'Ivoire", marker='o', markersize=12)
plt.plot(annee, e_burkina, color ='g', marker='o', markersize=6, linestyle ='dashed', linewidth = 2, label ='Burkina')
  
# Noms des axes
plt.xlabel('Années')
plt.ylabel("Consommation d'énergie en kwh")
  
# Titre
plt.title("Consommation d'électricité par habitant")
  
plt.legend()
plt.show()
```
